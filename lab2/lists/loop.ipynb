```python
thislist = ["apple", "banana", "cherry"] # Using for loop
for x in thislist:
  print(x)
```

```python
thislist = ["apple", "banana", "cherry"] # Using while loop
i = 0
while i < len(thislist):
  print(thislist[i])
  i = i + 1
```

```python
thislist = ["apple", "banana", "cherry"]
[print(x) for x in thislist]
```
